## Perguntas de exploração — Feature 1

```python
# 1. Volume por tabela
tabelas = {
    "customers": df_customers,
    "geolocation": df_geolocation,
    "order_items": df_order_items,
    "order_payments": df_order_payments,
    "order_reviews": df_order_reviews,
    "orders": df_orders,
    "products": df_products,
    "sellers": df_sellers,
}
for nome, df in tabelas.items():
    print(f"{nome}: {len(df):,} linhas, {df.shape[1]} colunas")
```

```python
# 2. Nulos por tabela
for nome, df in tabelas.items():
    nulos = df.isnull().sum()
    nulos = nulos[nulos > 0]
    if len(nulos) > 0:
        print(f"\n{nome}:")
        print(nulos)
    else:
        print(f"\n{nome}: sem nulos")
```

```python
# 3. Período coberto — orders
df_orders["order_purchase_timestamp"] = pd.to_datetime(df_orders["order_purchase_timestamp"])
print("Período dos pedidos:")
print("  Início:", df_orders["order_purchase_timestamp"].min())
print("  Fim:   ", df_orders["order_purchase_timestamp"].max())
print("  Pedidos por ano:")
print(df_orders["order_purchase_timestamp"].dt.year.value_counts().sort_index())
```

```python
# 4. Geocodificação — o que temos disponível para chegar ao código IBGE?
print("=== customers ===")
print(df_customers[["customer_zip_code_prefix", "customer_city", "customer_state"]].head(3))
print("\n=== sellers ===")
print(df_sellers[["seller_zip_code_prefix", "seller_city", "seller_state"]].head(3))
print("\n=== geolocation — colunas ===")
print(df_geolocation.columns.tolist())
print(df_geolocation.head(3))
print("\nZip prefixes distintos em customers:", df_customers["customer_zip_code_prefix"].nunique())
print("Zip prefixes distintos em geolocation:", df_geolocation["geolocation_zip_code_prefix"].nunique())
```

```python
# 5. Cobertura da geocodificação — quantos customers têm zip coberto pela geolocation?
zips_geo = set(df_geolocation["geolocation_zip_code_prefix"].unique())
zips_customers = set(df_customers["customer_zip_code_prefix"].unique())
cobertos = zips_customers & zips_geo
print(f"Zip prefixes em customers: {len(zips_customers)}")
print(f"Zip prefixes em geolocation: {len(zips_geo)}")
print(f"Cobertos: {len(cobertos)} ({len(cobertos)/len(zips_customers)*100:.1f}%)")
print(f"Não cobertos: {len(zips_customers - zips_geo)}")

# Quantos pedidos ficam sem cobertura?
orders_com_customer = df_orders.merge(df_customers[["customer_id", "customer_zip_code_prefix"]], on="customer_id")
sem_geo = orders_com_customer[~orders_com_customer["customer_zip_code_prefix"].isin(zips_geo)]
print(f"\nPedidos sem cobertura geo: {len(sem_geo)} ({len(sem_geo)/len(df_orders)*100:.1f}%)")
```

```python
# 6. Status dos pedidos — distribuição
print(df_orders["order_status"].value_counts())
print(f"\nTotal pedidos: {len(df_orders):,}")
print(f"Pedidos delivered: {(df_orders['order_status'] == 'delivered').sum():,}")
```
