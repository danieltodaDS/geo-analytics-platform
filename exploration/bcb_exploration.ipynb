# BCB PIX — adoção de pagamento digital por município

```python

# Libs
import time
import requests
import pandas as pd

```

```python
# Request

# DataBase se refere a data de inicio (offset), capturando todos os dados a partir dessa data
url = (
    "https://olinda.bcb.gov.br/olinda/servico/Pix_DadosAbertos/versao/v1/odata"
    "/TransacoesPixPorMunicipio(DataBase='201601')"
)

params = {"$format": "json"}

r = requests.get(url, params, timeout = 120)
```

```python
r.raise_for_status()
```

```python

data = r.json()
data['value'][:2]
```

```python
data['value'][1]
```

```python
from pydantic import BaseModel, field_validator

class PixMunicipioRaw(BaseModel):
    AnoMes: int
    Municipio_Ibge: int
    Municipio: str
    Estado_Ibge: int
    Estado: str
    Sigla_Regiao: str
    Regiao: str
    VL_PagadorPF: float
    QT_PagadorPF: int
    VL_PagadorPJ: float
    QT_PagadorPJ: int
    VL_RecebedorPF: float
    QT_RecebedorPF: int
    VL_RecebedorPJ: float
    QT_RecebedorPJ: int
    QT_PES_PagadorPF: int
    QT_PES_PagadorPJ: int
    QT_PES_RecebedorPF: int
    QT_PES_RecebedorPJ: int

    @field_validator("Municipio_Ibge", "Estado_Ibge", mode="before")
    @classmethod
    def cast_float_to_int(cls, v):
        # Pré-condição: v não é None — registros nulos são filtrados antes de chegar aqui
        return int(v)
```

```python

# Dataframe para facilitar a analise

# Min AnoMes = 202011 --> quando o Pix foi lançado
# Max AnoMes = 202606 --> mes atual
df = pd.DataFrame(data["value"]).reset_index(drop=True)
df.head()
```

```python

df["AnoMes"].value_counts().reset_index().sort_values(by='AnoMes')
df.shape

```

## Perguntas de exploração — Feature 3

```python
# 1. Paginação — a API retorna tudo em uma chamada ou há @odata.nextLink?
print("@odata.nextLink presente:", "@odata.nextLink" in data)
print("@odata.count presente:", "@odata.count" in data)
print("Total registros retornados:", len(data["value"]))
```

```python
# 2. Cobertura municipal — quantos municípios distintos? Todos os 5.571?
print("Municípios distintos na base:", df["Municipio_Ibge"].nunique())
print("Municípios esperados (IBGE):", 5571)
print("Diferença:", 5571 - df["Municipio_Ibge"].nunique())
```

```python
# 3. Lag de publicação — qual o AnoMes mais recente disponível?
print("AnoMes mais recente:", df["AnoMes"].max())
print("AnoMes mais antigo:", df["AnoMes"].min())
print("Meses distintos:", df["AnoMes"].nunique())

# Quantos municípios têm dado no mês mais recente?
mes_recente = df["AnoMes"].max()
cobertura_recente = df[df["AnoMes"] == mes_recente]["Municipio_Ibge"].nunique()
print(f"Municípios com dado em {mes_recente}:", cobertura_recente)
```

```python
# 4. Nulos — quais colunas têm valores nulos?
nulos = df.isnull().sum()
print("Nulos por coluna:")
print(nulos[nulos > 0] if nulos.sum() > 0 else "Nenhum nulo encontrado")
```

apesar do numero estar abaixo do esperado pode ter sido criado novos municipios

```python
# 5. Municípios sem cobertura em algum mês — há gaps?
# Esperado: 5571 municípios × n meses
total_meses = df["AnoMes"].nunique()
total_municipios = df["Municipio_Ibge"].nunique()
combinacoes_esperadas = total_meses * total_municipios
combinacoes_reais = len(df)
print(f"Meses: {total_meses}, Municípios: {total_municipios}")
print(f"Combinações esperadas (sem gaps): {combinacoes_esperadas}")
print(f"Combinações reais: {combinacoes_reais}")
print(f"Registros faltando: {combinacoes_esperadas - combinacoes_reais}")
```
